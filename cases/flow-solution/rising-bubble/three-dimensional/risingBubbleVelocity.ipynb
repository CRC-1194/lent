```python
idx = pd.IndexSlice
rcParams["figure.dpi"] = 300    
def plot_risingBubbleVelocity(dframe,viscosity,exp_value, Mo):
    """Plots the temporal evolution of the L-INFINITY velocity error norm for each mesh resolution and density ratio."""
    
    dframe=dframe.loc[idx[viscosity]]
    
    g=9.80665
    D=0.01305*2

    plt.plot(dframe["time"]/(D/g)**0.5, dframe["average velocity magnitude"]/(g*D)**0.5 )
   # plt.plot(dframe["time"]/(D/g)**0.5, exp_value )
    plt.axhline(y = exp_value, linestyle = ':')
    plt.title("N64,Mo=%s"%(Mo))
    plt.ylabel("Normalized buble rising velocity")
    plt.xlabel("Normalized time")
  # plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
         # fancybox=True, shadow=True, ncol=1)
```
